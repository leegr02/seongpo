6. 위 csv 파일에서 해결하고 싶은 질문은 무엇인지 구체적이고 명확하게 작성하시오.

: 2023년 상반기 자전거를 가장 많이 대여한 날과 가장 적게 대여한 날은? 그리고 각각의 대여건수는?

7. 위 질문을 해결하기 위해 csv 파일에서 출력하고 싶은 데이터 값을 구해보자. (예) 최대값, 최소값, 카운트, 합계, 평균

```python
import csv
f = open('20101_bicycle.csv')
data = csv.reader(f)
next(data)

max_bi = -1
max_date = ''
min_bi = 999999
min_date = ''

for row in data :
    #print(row)
    row[1] = row[1].replace(',','')
    row[1] = int(row[1])

    if max_bi < row[1] :
        max_date = row[0]
        max_bi = row[1]

    if min_bi > row[1] :
        min_date = row[0]
        min_bi = row[1]

print(f'2023년 상반기\n자전거를 가장 많이 대여한 날은 {max_date}이고, 대여건수는 {max_bi}이다.\n자전거를 가장 적게 대여한 날은 {min_date}이고, 대여건수는 {min_bi}이다.')
```

8. 그래프로 나타내고 싶은 속성을 새로운 리스트에 저장하고, 꺾은선 그래프로 나타내보자. (차트제목, 색 바꾸기)

```python
import csv
import matplotlib.pyplot as plt

f = open('20101_bicycle.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[1] != '':
        result.append(int(row[1].replace(',', '')))

print(result)

plt.rc('font', family = 'Malgun Gothic')
plt.title('2023년 상반기 서울시 공공자전거 대여건수')
plt.plot(result, 'r')
plt.show()
```

10. 그래프로 나타내고 싶은 속성을 두 가지 이상의 꺾은 선 그래프로 나타내보자.(조건을 세분화하여 새로운 리스트 2개 이상에 저장하고, 꺾은선 그래프로 나타낸다.) (차트제목, 색 바꾸기, 범례표시)

```python
#10번 코드
import csv
import matplotlib.pyplot as plt

f = open('20101_bicycle.csv')
data = csv.reader(f)
next(data)
spring = []
winter = []

for row in data :
    if row[1] != '':
        num = int(row[1].replace(',', ''))
        month = int(row[0].split('-')[1])
        if (month >= 4 and month <=5) :
            spring.append(num)
        elif (month >= 1 and month <= 2) :
            winter.append(num)

print(spring)
print(winter)

plt.rc('font', family = 'Malgun Gothic')
plt.title('자전거 대여 건수')
plt.plot(spring, 'yellowgreen', label='봄')
plt.plot(winter, 'gray', label='겨울')
plt.legend(loc = 5)
plt.show()
```

11. 10의 그래프와 같이 두 개의 꺾은 선 그래프를 그린 이유를 작성하시오.  
(날씨가 따뜻해질수록 자전거 대여건수가 증가하는 경향을 8번의 그래프에서 확인했다. 이를 바탕으로 봄(4,5월), 겨울(1,2월)로 대여건수를 나누어 꺾은선 그래프로 비교하고자 한다.)
